```python
from os import listdir
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from mpl_toolkits.axes_grid1 import ImageGrid
import numpy as np
import cv2
```

First create the collage for the original images

```python
import pandas as pd
pokemon = pd.read_csv('Original data.csv')
pokemon.pokemon = pokemon.pokemon.str.lower()
first_150 = pokemon.pokemon[0:150]
first_150[28] = 'nidoran-f'
first_150[31] = 'nidoran-m'
first_150[82] = 'farfetchd'
first_150[121] = 'mr-mime'
first_150 = first_150 + '.jpg'
```

```python
#first150 = listdir('images')
images = []
for pokemon in first_150:
    img = cv2.imread('images/' + pokemon)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    res = cv2.resize(img_rgb, dsize=(250,250))
    images.append(res)

fig = plt.figure(1, (30, 30))
grid = ImageGrid(fig, 110,
                 nrows_ncols=(15, 10),
                 axes_pad=.05,
)

for i in range(len(images)):
    grid[i].imshow(images[i])
    grid[i].axis('off')
plt.show(block=True)

fig.savefig('collage_original.jpg') # saving the image
```

Next, create the images for the main color images

```python
images = []
for pokemon in first_150:
    img = cv2.imread('new_images/' + pokemon)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    res = cv2.resize(img_rgb, dsize=(250,250))
    
    # fixing the background color on some of the images (some are not pure white)
    fixed_background = []
    for i in res:
        new_j = []
        for j in i:
            if np.all(j > 230):
                new_j.append([255,255,255])
            else:
                new_j.append(list(j))
        fixed_background.append(new_j)
        
    fixed_background = np.asarray(fixed_background)
    
    images.append(fixed_background)


fig = plt.figure(1, (30, 30))
grid = ImageGrid(fig, 110,
                 nrows_ncols=(15, 10),
                 axes_pad=.05,
)

for i in range(len(images)):
    grid[i].imshow(images[i])
    grid[i].axis('off')
plt.show(block=True)

fig.savefig('collage_new.jpg') # saving the image
```
